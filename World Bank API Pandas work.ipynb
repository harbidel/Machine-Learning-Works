# INDICATOR 1

```python
# Importing dataset from the word bank API base on the indicators collected from the Finance category
finance = wb.data.DataFrame(Indicator_ids_1, Country_codes,
                               columns='series', mrv=10) #time=range(2008, 2019)
finance.head()
```

# Renaming energy data index name 

```python
# resetting the dataset index
finance=finance.reset_index()

# renaming the index head
finance.rename(columns={'economy':'country',	'time':'year'}, inplace=True)

# setting back the index to default
finance.set_index(['country', 'year'], inplace=True)
finance.head()
```

```python
# renaming energy data index name 
finance.rename(columns={'FB.ATM.TOTL.P5':'ATM', 'FM.LBL.BMNY.GD.ZS':'BMNY', 'IC.CRD.INFO.XQ':'CRD', 'BX.KLT.DINV.CD.WD':'DINV'}, 
                      inplace=True)
finance.head()
```

#INDICATOR 2

```python
# Importing dataset from the word bank API base on the indicators collected from the Education category
education = wb.data.DataFrame(Indicator_ids_2, Country_codes,
                               columns='series', mrv=10) 
education.head()
```

# RESETTING INDEX

```python
# resetting the dataset index
education=education.reset_index()

# renaming the index head
education.rename(columns={'economy':'country',	'time':'year'}, inplace=True)

# setting back the index to default
education.set_index(['country', 'year'], inplace=True)
education.head()
```

```python
# renaming energy data index name 
education.rename(columns={'SE.PRM.UNER.FE':'FE', 'SE.PRM.UNER.MA':'MA', 'SE.ADT.LITR.MA.ZS':'MAZS', 'SE.PRM.PRSL.FE.ZS':'FEZS'}, 
                      inplace=True)
education.head()
```

# **3. DATA ANALYSIS**

# The minimum aggregate values for each on the countries

```python
# The minimum aggregate values for each on the countries
education.groupby('country').aggregate('min')
```

# The minimum sum aggregate values for each on the countries

```python
# The minimum sum aggregate values for each on the countries
finance.groupby('country').aggregate('min').sum()
```

# Using groupby with mean function 

```python
education.groupby('country')["MAZS","FEZS"].mean()
```

# Applying groupby with sum function 

```python
# applying groupby with sum function 
education.groupby('year').sum()
```

# Applying groupby with filter function 

```python
# applying groupby with filter function 
finance.groupby('country').filter(lambda x: x['ATM'].mean()>100)
```

# Applying groupby with apply function 

```python
# applying groupby with apply function 
group = education.groupby('country')
group['MAZS'].apply(lambda x: x.astype(float).max() - x.min())
```

# Applying groupby function with transfrom function

```python
# applying groupby function with transfrom function
finance[finance.groupby(['country'])['BMNY'].transform(sum) > 500]
```

# Using pivot_table function

```python
# pivot_table function
table = pd.pivot_table(data=education,index=['year'])
table.head()
```

# Using get_group() function with groupby() function to call all the data for year 2020

```python
# using get_group() function with groupby() function to call all the data for year 2020
finance.groupby(["year"]).get_group("YR2020")
```

# Using get_group() function with groupby() function to call all the data for year 2012

```python
# using get_group() function with groupby() function to call all the data for year 2012
education.groupby(["year"]).get_group("YR2012")
```
